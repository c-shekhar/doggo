# Image Feature Engineering

[OpenCV] using features from OpenCV, I try to get more information from each image, like the number of colors, the kinds of colors, etc. There's a lot of other features like corners, texture, edges we can engineer purely from image and use those as features to train classical machine learning classification models.

[OpenCV]: https://docs.opencv.org/master/d7/da8/tutorial_table_of_content_imgproc.html

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
from PIL import Image
def get_data_from_image(image_path):
    img = Image.open(image_path)
    cv_img = cv2.imread(image_path)
    plt.imshow(img)
    img_size = [img.size[0], img.size[1]]
    (means, stds) = cv2.meanStdDev(cv_img)
    mean_color = np.mean(cv_img.flatten())
    std_color = np.std(cv_img.flatten())
    color_stats = np.concatenate([means, stds]).flatten()
    df = pd.DataFrame([img_size + [mean_color] + [std_color] + color_stats.tolist()],
                      columns = ['img_size_x', 'img_size_y', 'img_mean_color', 'img_std_color', 'img_blue_mean', 'img_green_mean', 'img_red_mean', 'img_blue_std', 'image_green_std', 'image_red_std'])
    return df
```
